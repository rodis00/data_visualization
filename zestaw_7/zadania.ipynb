```python
def foo(liczba):
    suma = 0
    while liczba > 0:
        suma += liczba % 10
        liczba //= 10
    return suma
```

```python
foo(111)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('dosw7.csv', sep=';')

plt.title('Wyniki dowiadczenia')
plt.plot(df['Czas'], df['Zmienna1'], 'b--', label='zmienna 1')
plt.plot(df['Czas'], df['Zmienna2'], 'g', label='zmienna 2')
plt.xlabel('Czas')
plt.ylabel('Zmienna')
plt.legend(loc='lower left')
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('sport6.csv', sep='_')

m = df[df['Pe'] == 'M']
k = df[df['Pe'] == 'K']

color1 = [np.random.rand(3) for i in range(len(m.index))]
color2 = [np.random.rand(3) for i in range(len(k.index))]

plt.subplot(2, 1, 1)
plt.title('Mczyni')
plt.bar(m['Sport'], m['Popularno'], color=color1)
plt.ylim(0, 60)

plt.subplot(2, 1, 2)
plt.title('Kobiety')
plt.bar(k['Sport'], k['Popularno'], color=color2)
plt.ylim(0, 60)
plt.tight_layout()
plt.show()
```
