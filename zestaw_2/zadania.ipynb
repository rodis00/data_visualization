```python
def foo(*kwargs):
    list = []
    for kwarg in kwargs:
        if kwarg not in list:
            list.append(kwarg)
    return list
```

```python
l1 = [1, 2, 3, 2]
foo(l1, 'a', 'abc', 'a')
```

```python
def foo(number):
    suma = 0
    while number > 0:
        a = number % 10
        number //= 10
        suma += a
    return suma
```

```python
number = 34512
foo(number)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
### zad_4
```

```python
df = pd.read_csv('dosw.csv')

czas = df['Czas']
val_1 = df['Zmienna1']
val_2 = df['Zmienna2']
```

```python
plt.title('Wykres liniowy wartosci 1 i 2 wzgledem czasu')
plt.plot(czas, val_1, '--', color='brown', label='Zmienna1')
plt.plot(czas, val_2, color='red', label='Zmienna2')
plt.xlabel('Czas')
plt.ylabel('Wartości')
plt.legend(loc='center left')
plt.grid()
plt.xticks([2, 3, 4, 5])
plt.yticks([-15, -10, -5])
plt.savefig('zad_4.png')
plt.show()
```

```python
df = pd.read_csv('medale.csv', sep=';')

suma = df['Złote'] + df['Srebrne'] + df['Brązowe']
letnie = suma[df["Rodzaj"] == "Letnie"]
zimowe = suma[df["Rodzaj"] == "Zimowe"]
```

```python
x = np.arange(len(letnie))

lato = df[df['Rodzaj'] == 'Letnie']
zima = df[df['Rodzaj'] == 'Zimowe']

plt.subplot(2, 1, 1)
plt.bar(x, letnie, color="orange")
plt.xticks(x, lato['Miejsce'])
plt.yticks([0, 2, 4, 6, 8, 10, 12, 14])
plt.title("Medale Polski - letnie olimpiady")
plt.subplot(2, 1, 2)
plt.bar(x, zimowe)
plt.xticks(x, zima['Miejsce'])
plt.title("Medale Polski - zimowe olimpiady")
plt.tight_layout()
plt.show()
```
