```python
def mnozenie(*args):
    wynik = 1
    for el in args:
        wynik *= el
    return wynik
```

```python
mnozenie(2, 2, 4)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('planety.csv', sep=',')

planety = df[df['Masa'] > 100000]

color = [np.random.rand(3) for x in range(len(planety.index))]

plt.title('Masa planet')
plt.pie(planety['Masa'], labels=planety['Planeta'], autopct='%.1f%%', colors=color)
plt.annotate('12345', (-1.25, -1))
plt.savefig('zad_4.png')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('dosw11.csv', sep=';')

plt.subplot(2, 1, 1)
plt.title('Zmienna od czasu')
plt.plot(df['Czas'], df['Zmienna'], label='zmienna', color='r')
plt.legend(loc='upper left')

plt.subplot(2, 1, 2)
plt.title('Zmienna od czasu 2')

higher = df[df['Zmienna'] > 10]

higher_value = higher['Zmienna'].values

df['Zmienna'] = df['Zmienna'].replace(higher_value, 1.13)

plt.plot(df['Czas'], df['Zmienna'], label='zmienna2', color='g')
plt.legend(loc='upper right')
plt.tight_layout()
plt.show()
```
