```python
def foo(lista):
    wyraz = ''
    dlugosc = len(lista[0])
    for el in lista:
        if dlugosc > len(el):
            dlugosc = len(el)
            wyraz = el
    return wyraz
```

```python
foo(['Ala', 'ma', 'kota'])
```

```python
class Wektor:
    a = 0
    b = 0

    def dodawanie(w1, w2):
        w = Wektor()
        w.a = w1.a + w2.a
        w.b = w1.b + w2.b
        return w
```

```python
w1 = Wektor()
w1.a = 1
w1.b = 3

w2 = Wektor()
w2.a = 2
w2.b = 4

w3 = Wektor.dodawanie(w1, w2)
```

```python
w3.a, w3.b
```

```python
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('rocky.csv')

rgb = [np.random.rand(3) for x in range(len(df.index))]
color = ["#"+''.join([random.choice('0123456789ABCDEF') for j in range(6)]) for x in range(len(df.index))]
```

```python
color
```

```python
plt.title('Zyski')
plt.barh(df['Nazwa'], df['Zyski w mln $'], color=color) # or use color=rgb
plt.xlabel('Nazwa')
plt.ylabel('Zyski w mln $')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('sport5.csv', sep='#')

k = df[df['Pe'] == 'K']
m = df[df['Pe'] == 'M']

plt.subplot(1, 2, 1)
plt.title('Zainteresowanie sportem mczyzn')
plt.pie(m['Popularno'], labels=m['Sport'], autopct='%.0f%%')

plt.subplot(1, 2, 2)
plt.title('Zainteresowanie sportem kobiet')
plt.pie(k['Popularno'], labels=k['Sport'], autopct='%.0f%%')
plt.tight_layout()
plt.show()
```
